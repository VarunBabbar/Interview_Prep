```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def levelOrderBottom(self, root: TreeNode) # -> list[list[int]]: 
    ''' Here I tried a new method which employs a dictionary to recursively
        store all the nodal values at each level. 
        
        Time Complexity: O(n+l) where n is the number of nodes and l is the number of levels
        Space Complexity: O(n+l) because we create a dictionary of size l and the total 
        number of elements in the dictionary is n
        '''
       # 
        #       For Example: 
        #  The dictionary looks like:
        # dict = {0: [4], 1: [5,7], 2: [1,2,8,9]}
        
        #     The Corresponding Tree is :
        ###              4
        ###             / \ 
        ###            /   \
        ###           /     \
        ##           /       \
        ##           5        7
        ###        /   \     / \
        ###       /     \   8  9
        ##       /       \
        #       1        2 
           
        if root is None:
            return []
        level_dict = {}
        level_dict[0] = [root.val] # Adding the root node value at level 0 
        level = 1
        level_dict = self.recurse(root,level_dict,level) 
        vals = [level_dict[i] for i in range(len(level_dict)-1,-1,-1)]
        return vals
    def recurse(self,node,level_dict,level): 
        if node.left is not None: 
            try:
                # If there is already a node at this level, then add this node to it 
                level_dict[level].append(node.left.val) 
            except:
                # Otherwise index a new array into the dictionary
                level_dict[level] = [node.left.val]
            level_dict = self.recurse(node.left,level_dict,level+1) 
            # Traversing up the tree recursively
        if node.right is not None:
            try:
                # If there is already a node at this level, then add this node to it 
                level_dict[level].append(node.right.val) 
            except:
                # Otherwise index a new array into the dictionary
                level_dict[level] = [node.right.val]
            
            level_dict = self.recurse(node.right,level_dict,level+1)
             # Traversing up the tree recursively
        return level_dict     
# Input a TreeNode object to this class
```
